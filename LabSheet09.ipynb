## Lab sheet 9: Machine Translation using Transformers

<b>Aim</b>: To implement a transformer-based machine translation model using libraries like HuggingFace or Fairseq, and evaluate the translation quality using BLEU score.  

```python
from transformers import MarianMTModel, MarianTokenizer
import sacrebleu
```

```python
model_name = "Helsinki-NLP/opus-mt-en-hi"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name, use_safetensors=True)
```

```python
def translate(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
input_text = "natural language processing is simple"
translated_text = translate(input_text)
print("Input:", input_text)
print("Translated:", translated_text)
```

```python
reference = ["प्राकृतिक भाषा प्रक्रिया सरल है"]
candidate = [translated_text]
bleu = sacrebleu.corpus_bleu(candidate, [reference])
print("BLEU Score:", bleu.score)
```
